```python
import pandas as pd

# Example DataFrame setup
data = {
    "Algorithm": ["A*", "BFS", "DFS", "Dijkstra"],
    "Time": [1.2, 0.8, 0.5, 1.1],  # hypothetical times
    "Memory": [50, 30, 20, 40]  # hypothetical memory usage in MB
}
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# Set up the figure and axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting time comparisons
time_bars = ax1.bar(df['Algorithm'], df['Time'], color='b', label='Time (seconds)')
ax1.set_xlabel('Algorithm')
ax1.set_ylabel('Execution Time (seconds)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a twin axis for memory
ax2 = ax1.twinx()
memory_bars = ax2.bar(df['Algorithm'], df['Memory'], color='r', alpha=0.7, label='Memory (MB)')
ax2.set_ylabel('Memory Usage (MB)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Adding legends
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.title('Comparison of Algorithm Performance')
plt.show()
```
